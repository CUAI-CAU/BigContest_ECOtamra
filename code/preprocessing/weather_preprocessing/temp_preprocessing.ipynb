```python
import pandas as pd
import numpy as np
```

```python
data_temp = pd.read_csv("./기온/한림읍_기온_202106_202106.csv", encoding='cp949')
data_temp
```

```python
data_temp = data_temp.drop([720], 0) # 24시간 x 30일

data_temp.columns=["day","hour","temp"]
data_temp.insert(0,'month',"")
data_temp.insert(0,'year',"")
data_temp
```

```python
import warnings
warnings.filterwarnings('ignore')

# month 채우기(2020년은 바꿔야함)
month='06'
data_temp['month']=month

# year 채우기
year='2021'
data_temp['year']=year


# base_date(2021-06-day) 형식으로 정리
data_temp.insert(0,'base_date',"")
for i in range(720):
    data_temp['base_date'][i]=data_temp['year'][i]+'-'+data_temp['month'][i]+'-'+data_temp['day'][i]

# 매달 사이의 설명행 제거
data_temp=data_temp.dropna(axis=0)

data_temp
```

```python
# 불필요한 행 삭제
data_temp =data_temp.drop(['year','month','day', 'hour'],1)
data_temp
```

```python
data_temp['base_date'] = pd.to_datetime(data_temp['base_date'], format="%Y-%m-%d")
data_temp_res = data_temp.resample('M', on='base_date').mean() 
data_temp_res
```

```python
data_temp_res.to_csv("files\\data_preprocessing\\한림읍_2018.csv", index_label=False)
pd.read_csv("files\\data_preprocessing\\한림읍_2018.csv")
```
