```python
import pandas as pd
import numpy as np
```

```python
Card_spend=pd.read_csv("./04_음식관련 카드소비_CARD_SPENDING.CSV", encoding='cp949')
```

```python
Card_spend.head(10)
```

```python
# 필요없는 열 삭제 (날짜, 행정동코드, 행정동명, 결제건수(use_cnt), 결제금액(use_amt))
Card_spend_1=Card_spend.drop(['city','emd_nm','mct_cat_cd', 'mct_cat_nm'], 1)
Card_spend_1
```

```python
# 날짜데이터 datetime으로 변환
Card_spend_1['base_date'] = pd.to_datetime(Card_spend_1['base_date'], format="%Y-%m-%d")
Card_spend_1.set_index(Card_spend_1['base_date'], inplace=True)
# "알수없음" 삭제
idx = Card_spend_1[Card_spend_1['emd_cd'] == "XXXXXXXX"].index
Card_spend_1 = Card_spend_1.drop(idx)

# 행정동코드 int형으로 변환
Card_spend_1['emd_cd'] = pd.to_numeric(Card_spend_1['emd_cd'])

# Card_spend_1.dtypes
# base_date    datetime64[ns]
# emd_cd                int64
# use_cnt               int64
# use_amt               int6

Card_spend_1.head(30)
```

```python
# 행정동코드 매개변수로 받아서 해당 구역의 배출량을 월별로 정리

pd.options.display.float_format = '{:.5f}'.format
def monthly_output(emdcd):
    emd_data = Card_spend_1[Card_spend_1['emd_cd'] == emdcd]    # 해당 행정동 데이터만 추출
    emd_data_monthly = emd_data.resample('M', on='base_date').mean()    # 월별로 합산
    emd_data_monthly['emd_cd'] = emdcd
    return emd_data_monthly
```

```python
emd_cd_ary = [50110250, 50110253, 50110256, 50110259, 50110310, 50110320, 50110330, 50110510, 50110520, 50110530, 50110540, 50110550, 50110560, 50110570, 50110580, 50110590, 50110600, 50110610, 50110620, 50110630, 50110640, 50110650, 50110660, 50110670, 50110680, 50110690, 50130250, 50130253, 50130259, 50130310, 50130320, 50130510, 50130520, 50130530, 50130540, 50130550,
              50130560, 50130570, 50130580, 50130590, 50130600, 50130610, 50130620]

data_result=pd.DataFrame(index=range(0,0), columns=['base_date','emd_cd','use_cnt','use_amt'])

for i in emd_cd_ary:
    tmp = monthly_output(i)
    data_result = data_result.append(tmp)

data_result['base_date'] = data_result.index

data_result
```

```python
data_result.to_csv("./데이터 전처리/Card_spending.csv",index_label=False)
```
