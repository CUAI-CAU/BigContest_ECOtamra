```python
date7 = pd.date_range(start='2021-02-15', end='2021-05-29')
data7 = pd.DataFrame({'base_date':date7})
data7['distancing'] = 1.5
data7.head()
```

```python
date8 = pd.date_range(start='2021-05-30', end='2021-06-30')
data8 = pd.DataFrame({'base_date':date8})
data8['distancing'] = 2
data8.head()
```

```python
distancing_data = data1.append(data2)
distancing_data = distancing_data.append(data3)
distancing_data = distancing_data.append(data4)
distancing_data = distancing_data.append(data5)
distancing_data = distancing_data.append(data6)
distancing_data = distancing_data.append(data7)
distancing_data = distancing_data.append(data8)
distancing_data
```

```python
distancing_data_monthly = distancing_data.resample('M', on='base_date').mean()
distancing_data_monthly.head()
```

```python
distancing_data_monthly['distancing']['2021-05-31']
```

```python
def monthly_output(date, emdcd):
    df = pd.DataFrame(index=range(0,1), columns=['base_date', 'emd_cd', 'distancing'])
    df['base_date'] = date
    df['emd_cd'] = emdcd
    df['distancing'] = distancing_data_monthly['distancing'][date]
    return df
```

```python
emd_cd_ary = [50110250, 50110253, 50110256, 50110259, 50110310, 50110320, 50110330, 50110510, 50110520, 50110530, 50110540, 50110550, 50110560, 50110570, 50110580, 50110590, 50110600, 50110610, 50110620, 50110630, 50110640, 50110650, 50110660, 50110670, 50110680, 50110690, 50130250, 50130253, 50130259, 50130310, 50130320, 50130510, 50130520, 50130530, 50130540, 50130550, 50130560, 50130570, 50130580, 50130590, 50130600, 50130610, 50130620]
distancing_result = pd.DataFrame(index=range(0,0), columns=['base_date', 'emd_cd', 'distancing'])

for i in distancing_data_monthly.index:
    for j in emd_cd_ary:
        tmp = monthly_output(i, j)
        distancing_result = distancing_result.append(tmp)
        
distancing_result = distancing_result.reset_index(drop=True)
distancing_result
```
