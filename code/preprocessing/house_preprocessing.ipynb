```python
import pandas as pd
import numpy as np
```

```python
data20 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data\\house_2020.csv", encoding='cp949')
data20.head()
```

```python
data20.columns = ['emd_nm', 'total_house', 'detached', 'apt', 'town', 'multiplex', 'commercial_building', 'emd_cd']
data20['base_date'] = "2020-01-31"
data20['base_date'] = pd.to_datetime(data20['base_date'], format="%Y-%m-%d")
data20 = data20[['base_date', 'emd_nm', 'emd_cd', 'total_house', 'detached', 'apt', 'town', 'multiplex', 'commercial_building']]
data20
```

```python
month = ['2020-02-29', '2020-03-31', '2020-04-30', '2020-05-31', '2020-06-30', '2020-07-31', '2020-08-31',
        '2020-09-30', '2020-10-31', '2020-11-30', '2020-12-31', '2021-01-31', '2021-02-28', '2021-03-31',
        '2021-04-30', '2021-05-31', '2021-06-30']

data = data20.copy()
for i in month:
    tmp = data20.copy()
    tmp['base_date'] = i
    tmp['base_date'] = pd.to_datetime(tmp['base_date'], format="%Y-%m-%d")
    data = data.append(tmp)
data
```
