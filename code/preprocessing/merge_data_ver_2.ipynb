```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data_preprocessing\\1_food_waste_preprocessing.csv")
data2_1 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data_preprocessing\\2_1_korean_preprocessing_v2.csv")
data2_2 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data_preprocessing\\2_2_long_term_frgn_preprocessing_v2.csv")
data2_3 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data_preprocessing\\2_3_short_term_frgn_3_resident_pop_preprocessing.csv")
data3 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data_preprocessing\\resident_pop_processing2(sex).csv")
data4 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data_preprocessing\\4_card_spending_ver2.csv")
```
