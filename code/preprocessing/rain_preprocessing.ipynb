```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("./데이터 전처리/강수량 데이터/2018(1~12)/한림읍_강수_201801_201812.csv", encoding='cp949')
data1
```

```python
data1.columns=["day","hour","rain_amt"]
data1.insert(0,'month',"")
data1.insert(0,'year',"")
data1
```

```python
import warnings
warnings.filterwarnings('ignore')
# month 채우기(2020년은 바꿔야함)
data1.at[0:744,'month']='01' #1월 (월 사이마다 부연설명 행1개 있음)
data1.at[745:1417,'month']='02' #2월
data1.at[1418:2162,'month']='03'# 3월
data1.at[2163:2883,'month']='04'
data1.at[2884:3628,'month']='05'
data1.at[3629:4349,'month']='06'
data1.at[4350:5094,'month']='07'
data1.at[5095:5839,'month']='08'
data1.at[5840:6560,'month']='09'
data1.at[6561:7305,'month']='10'
data1.at[7306:8026,'month']='11'
data1.at[8027:8771,'month']='12'

# year 채우기
year='2018'
data1['year']=year


# base_date(2018-0-0) 형식으로 정리
data1.insert(0,'base_date',"")
for i in range(8771):
    data1['base_date'][i]=data1['year'][i]+'-'+data1['month'][i]+'-'+data1['day'][i]

# 매달 사이의 설명행 제거
data1=data1.dropna(axis=0)
```

```python
# 불필요한 행 삭제
data1_drop=data1.drop(['year','month','day'],1)
data1_drop
```

```python
data1_drop['base_date'].unique()
data1_drop.head(40)
data1_grouped=data1_drop.groupby('base_date').sum()
data1_grouped.head(40)
data1_grouped=data1_grouped.drop('hour',1)
data1_grouped
# 날짜순 정렬 안됨
```
