```python
import pandas as pd
import numpy as np
```

```python
data2_1 = pd.read_csv("files\\data\\2_1_korean.csv", encoding='cp949')
data2_1.head()
```

```python
# 필요없는 열 삭제 (날짜, 행정동코드, 행정동명, 배출량만 남김)
data2_1_pre = data2_1.drop(['time', 'city', 'resd'], 1)
data2_1_pre
```

```python
# 각 열을 하나의 type으로 통일

# 날짜데이터 datetime으로 변환
data2_1_pre['base_date'] = pd.to_datetime(data2_1_pre['base_date'], format="%Y-%m-%d")

# "알수없음" 삭제
idx = data2_1_pre[data2_1_pre['emd_cd'] == "알수없음"].index
data2_1_pre = data2_1_pre.drop(idx)

# 행정동코드 int형으로 변환
data2_1_pre['emd_cd'] = pd.to_numeric(data2_1_pre['emd_cd'])
# 인구수 int형으로 변환
data2_1_pre['resd_pop_cnt'] = pd.to_numeric(data2_1_pre['resd_pop_cnt'])
data2_1_pre['work_pop_cnt'] = pd.to_numeric(data2_1_pre['work_pop_cnt'])
data2_1_pre['visit_pop_cnt'] = pd.to_numeric(data2_1_pre['visit_pop_cnt'])
# 나이 int형으로 변환
data2_1_pre['age'] = pd.to_numeric(data2_1_pre['age'])

data2_1_pre
```

```python
# 성별로 데이터 분할
groups = data2_1_pre.groupby(data2_1_pre.sex)
female_df = groups.get_group("여성")
male_df = groups.get_group("남성")
```

```python
female_df
```

```python
# 여성데이터 연령별로 분할
female_0 = female_df.groupby('age').get_group(0)
female_10 = female_df.groupby('age').get_group(10)
female_20 = female_df.groupby('age').get_group(20)
female_30 = female_df.groupby('age').get_group(30)
female_40 = female_df.groupby('age').get_group(40)
female_50 = female_df.groupby('age').get_group(50)
female_60 = female_df.groupby('age').get_group(60)
female_70 = female_df.groupby('age').get_group(70)
female_80 = female_df.groupby('age').get_group(80)
```

```python
# 남성데이터 연령별로 분할
male_0 = male_df.groupby('age').get_group(0)
male_10 = male_df.groupby('age').get_group(10)
male_20 = male_df.groupby('age').get_group(20)
male_30 = male_df.groupby('age').get_group(30)
male_40 = male_df.groupby('age').get_group(40)
male_50 = male_df.groupby('age').get_group(50)
male_60 = male_df.groupby('age').get_group(60)
male_70 = male_df.groupby('age').get_group(70)
male_80 = male_df.groupby('age').get_group(80)
```

```python
# 인구수 월별 합산 함수
def monthly_output(emdcd, df):
    emd_data = df[df['emd_cd'] == emdcd]    # 해당 행정동 데이터만 추출
    if len(emd_data != 0):
        emd_data_monthly = emd_data.resample('M', on='base_date').sum()    # 월별로 합산
        emd_data_monthly['emd_cd'] = emdcd
        return emd_data_monthly
    else:
        return
```

```python
# 행정동코드 배열
emd_cd_ary = [50110250, 50110253, 50110256, 50110259, 50110310, 50110320, 50110330, 50110510, 50110520, 50110530, 50110540, 50110550, 50110560, 50110570, 50110580, 50110590, 50110600, 50110610, 50110620, 50110630, 50110640, 50110650, 50110660, 50110670, 50110680, 50110690, 50130250, 50130253, 50130259, 50130310, 50130320, 50130510, 50130520, 50130530, 50130540, 50130550, 50130560, 50130570, 50130580, 50130590, 50130600, 50130610, 50130620]

# 여성 0대 데이터
fem_0_result = pd.DataFrame(index=range(0,0), columns=['base_date', 'emd_cd','resd_pop_cnt', 'work_pop_cnt', 'visit_pop_cnt'])

for i in emd_cd_ary:
    tmp = monthly_output(i, female_0)
    fem_0_result = fem_0_result.append(tmp)
    
fem_0_result['base_date'] = fem_0_result.index
fem_0_result = fem_0_result.drop(['age'], 1) # 연령변수 제거
fem_0_result.rename(columns={'resd_pop_cnt':'f0_resd_pop', 'work_pop_cnt':'f0_work_pop', 'visit_pop_cnt':'f0_visit_pop'}, inplace=True) # 변수명 수정
fem_0_result
```
