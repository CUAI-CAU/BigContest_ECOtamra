```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("files\\data\\weather_2019.csv", encoding='cp949')
data1.head()
```

```python
# 필요없는 열 삭제 
data1_pre = data1.drop(['시도명', '최대 풍속', '업종명', '이용금액', '데이터기준일자'], 1)
data1_pre
```

```python
#변수명 통일
data1_pre.rename(columns={'일자':'base_date', '읍면동명':'emd_nm', '평균 기온':'temp', '일강수량':'rain'}, inplace=True)
data1_pre
```

```python
data_emd = pd.read_csv("files\\data\\데이터정의서.csv", encoding='cp949')
data_emd.head()
```

```python
# 필요없는 열 삭제 
data_emd = data_emd.drop(['No', 'Unnamed: 3'], 1)

# 필요 없는 행 삭제 (알수없음)
data_emd = data_emd.drop(index = 43, axis = 0)

#변수명 통일
data_emd.rename(columns={'행정동코드':'emd_cd','행정동명':'emd_nm'}, inplace=True)
data_emd
```

```python
# 데이터 프레임 합치기
data1_emd = pd.merge(data1_pre, data_emd)

# 각 열을 하나의 type으로 통일
# 날짜데이터 datetime으로 변환
data1_emd['base_date'] = pd.to_datetime(data1_emd['base_date'], format="%Y-%m-%d")

# 인구수 int형으로 변환
data1_emd['temp'] = pd.to_numeric(data1_emd['temp'])
data1_emd['rain'] = pd.to_numeric(data1_emd['rain'])

# 행정동코드 int형으로 변환
data1_emd['emd_cd'] = pd.to_numeric(data1_emd['emd_cd'])

data1_emd
```

```python
# 행정동코드 매개변수로 받아서 해당 구역의 배출량을 월별로 정리
def monthly_output(emdcd):
    emd_data = data1_emd[data1_emd['emd_cd'] == emdcd]    # 해당 행정동 데이터만 추출
    if len(emd_data != 0):
        emd_data_monthly = emd_data.resample('M', on='base_date').sum()    # 월별로 합산
        emd_data_monthly['emd_cd'] = emdcd
        return emd_data_monthly
    else:
        return
```

```python
emd_cd_ary = [50110250, 50110253, 50110256, 50110259, 50110310, 50110320, 50110330, 50110510, 50110520, 50110530, 50110540, 50110550, 50110560, 50110570, 50110580, 50110590, 50110600, 50110610, 50110620, 50110630, 50110640, 50110650, 50110660, 50110670, 50110680, 50110690, 50130250, 50130253, 50130259, 50130310, 50130320, 50130510, 50130520, 50130530, 50130540, 50130550, 50130560, 50130570, 50130580, 50130590, 50130600, 50130610, 50130620]

data1_result = pd.DataFrame(index=range(0,0), columns=['base_date', 'emd_cd', 'temp', 'rain'])

for i in emd_cd_ary:
    tmp = monthly_output(i)
    data1_result = data1_result.append(tmp)

data1_result['base_date'] = data1_result.index
data1_result
```

```python
# 기온 월평균, 강수량은 월별합
data1_result['day'] = data1_result['base_date'].dt.day
data1_result['temp'] = data1_result['temp']/data1_result['day']
data1_result = data1_result.drop(['day'], 1)
data1_result                                                   
```
