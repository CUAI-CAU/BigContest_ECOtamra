```python
feature = data.drop(['em_g'], 1)

em_g = data[['em_g']]
x_card_cnt = data['use_cnt']

x_data1 = sm.add_constant(x_card_cnt, has_constant = "add")

multi_model = sm.OLS(em_g, x_card_cnt)
fitted_multi_model = multi_model.fit()

fitted_multi_model.summary()
```
