```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data\\1_food_waste.csv", encoding='cp949')
data4 = pd.read_csv("C:\\Users\\asus\\BigContest\\files\\data\\4_card_spending.csv", encoding='cp949')
```

# 음식물쓰레기배출량 데이터

```python
data1
```

```python
# "알수없음" 가져오기
idx = data1[data1['emd_nm'] == "알수없음"].index
data1_pre = data1.loc[idx]
data1_pre
```

```python
# 필요 없는 열 삭제
data1_pre = data1_pre.drop(['city', 'emd_cd', 'emd_nm', 'em_area_cd', 'pay_amt'], 1)
data1_pre
```

```python
# 날짜데이터 datetime으로 변환
data1_pre['base_date'] = pd.to_datetime(data1_pre['base_date'], format="%Y-%m-%d")
data1_pre.set_index(data1_pre['base_date'])

# 배출건수 int형으로 변환
data1_pre['em_cnt'] = pd.to_numeric(data1_pre['em_cnt'])

# 배출량 int형으로 변환
data1_pre['em_g'] = pd.to_numeric(data1_pre['em_g'])

data1_pre
```

```python
# 월별로 합산
data1_pre_monthly = data1_pre.resample('M', on='base_date').sum()
data1_pre_monthly
```

```python
data1_pre_monthly.to_csv("C:\\Users\\asus\\BigContest\\code\\unknown_waste_monthly.csv", index_label=False)
```
