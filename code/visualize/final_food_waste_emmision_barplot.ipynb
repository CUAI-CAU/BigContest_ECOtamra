```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("files\\data\\predict_2107_2108.csv", encoding='cp949')
df
```

```python
plt.rcParams["figure.figsize"] = (50,30)
df.plot.bar(x='emd_cd', y='2021-07-31', rot=0)
plt.title('Food Waste Emissions in July 2021', fontsize=45)
plt.xlabel('emd_cd', fontsize=20)
plt.ylabel('em_g', fontsize=20)
plt.legend(loc='best', ncol=2, fontsize=30)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (50,30)
df.plot.bar(x='emd_cd', y='2021-08-31', rot=0)
plt.title('Food Waste Emissions in August 2021', fontsize=45)
plt.xlabel('emd_cd', fontsize=20)
plt.ylabel('em_g', fontsize=20)
plt.legend(loc='best', ncol=2, fontsize=30)
plt.show()
```

```python
# 내림차순으로 정리
df = df.groupby('emd_cd').mean().sort_values(['2021-07-31'],ascending=False)
df.plot(kind='bar', figsize=(50,30))
plt.title('Ranking of Food Waste Emissions', fontsize=45)
plt.xlabel('emd_cd', fontsize=20)
plt.ylabel('em_g', fontsize=20)
plt.legend(loc='best', ncol=2, fontsize=30)
plt.show()
```
