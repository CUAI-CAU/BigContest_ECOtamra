```python
%matplotlib inline
import pandas as pd                # 데이터를 저장하고 처리하는 패키지
import matplotlib as mpl           # 그래프를 그리는 패키지
import matplotlib.pyplot as plt    # 그래프를 그리는 패키지
```

```python
df1 = pd.read_csv("files\\data\\predict_2107_2108.csv", encoding='cp949')
df1
```

```python
df7=df1.drop(['2021-08-31'],1)
df7.insert(0,'year',"2021-07-31")
df7.rename(columns={"2021-07-31": 'em_g', "year":'base_date'}, inplace=True)
df7['base_date'] = pd.to_datetime(df7['base_date'], format="%Y-%m-%d")
df7.set_index(df7['base_date'])
df7
```

```python
df8 = df1.drop(['2021-07-31'],1)
df8.insert(0,'year',"2021-08-31")
df8.rename(columns={"2021-08-31": 'em_g', "year":'base_date'}, inplace=True)
df8['base_date'] = pd.to_datetime(df8['base_date'], format="%Y-%m-%d")
df8.set_index(df7['base_date'])
df8
```

```python
res = pd.concat([df7,df8])
res
```

```python
df2 = pd.read_csv("files\\data_preprocessing\\food_waste_preprocessing.csv", encoding='cp949')
df2=df2.drop(['em_cnt'],1)
df2['base_date'] = pd.to_datetime(df2['base_date'], format="%Y-%m-%d")
df2
```

```python
result = pd.concat([df2, res])
result.set_index(result['base_date'])
result['base_date'] = pd.to_datetime(result['base_date'], format="%Y-%m-%d")
result
```

```python
data = result.resample('M', on='base_date').sum() 
df=data.drop(['emd_cd'],1)
df['base_date']= df.index
df
```

```python
fig = plt.figure(figsize=(10,10)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(df['base_date'],df['em_g']) ## 선그래프 생성
 
plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 40도 회전 
plt.title('Food Waste Emmision Graph(2018.01-2021.08)',fontsize=20) ## 타이틀 설정
plt.show()
```
